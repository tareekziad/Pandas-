# Bivariate / Multivariate Analysis

#### What is the most profited Segment ?

```python
df.groupby('Segment')['Profit'].sum().to_frame()
```

```python
sns.barplot(data = df , x = 'Segment' , y = 'Profit' , estimator = np.sum)
```

#### What is the top 10 profited state ?

```python
top10 = df.groupby('State')['Profit'].sum().sort_values(ascending = False).head(10).to_frame().reset_index()
```

```python
sns.barplot(data = top10 , x = top10.columns[0] , y = top10.columns[1])
```

#### What is the most discounted category ?

```python
top_cat = df.groupby('Category')['Discount'].sum().sort_values(ascending = False)
```

```python
sns.barplot(data = df , x = 'Category' , y = 'Discount' , 
            estimator = np.sum , order = top_cat.index)
```

### What is the total profit for each Category in each Region

```python
dt = df.groupby(['Region' , 'Category'])['Profit'].sum().to_frame().unstack()
dt
```

```python
sns.barplot(data = df , x = 'Region' , y = 'Profit' , hue = 'Category' 
            , estimator = np.sum , order = dt.index)
```

```python
piovt = df.pivot_table(columns = 'Region' , index = 'Category' , values = 'Profit' , aggfunc = np.sum)
```

```python
sns.heatmap(piovt , annot = True)
```

#### Extract Year and Month of order data

```python
df['Year'] = df['Order Date'].apply(lambda x : x.split('/')[2])
df['Year'].unique()
```

```python
df['Month'] = df['Order Date'].apply(lambda x : x.split('/')[0])
df['Month'].unique()
```

### what's is the amount of Discount every month ?

```python
dt = df.groupby('Month')['Discount'].sum().to_frame().reset_index().sort_values(by = 'Discount', ascending = False)
```

```python
sns.barplot(data = df , x = 'Month' , y = 'Discount' , estimator = np.sum , order = dt['Month'])
```

### what's is the amount of Discount every month each year ?

```python
df.groupby(['Year' , 'Month'])['Discount'].sum().to_frame().unstack()
```

```python
sns.barplot(data = df , x = 'Year' , y = 'Discount' , hue = 'Month' , estimator = np.sum)
```

```python
pivt = df.pivot_table(columns = 'Year' , index = 'Month' , values = 'Discount' , aggfunc = np.sum)
```

```python
sns.heatmap(pivt , annot = True)
```

### what's the effect of Discount on Sales and Profit(Lineplot)


```python
sns.lineplot(data = df , y = 'Sales' , x = 'Discount' , hue = 'Year' ,estimator = np.sum , ci = True)
```

```python
sns.lineplot(data = df , x = 'Discount' , y = 'Profit' , hue = 'Year' , estimator = np.sum , ci = True)
```

### what is the total sells for each region each year ?

```python
sns.barplot(data = df , x = 'Region' , y = 'Sales' , hue = 'Year' , estimator = np.sum , ci = True)
```

```python
pivt = df.pivot_table(columns = 'Region' , index = 'Year' , values = 'Sales' , aggfunc = np.sum)
```
